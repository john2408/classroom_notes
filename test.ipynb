```python
import pandas as pd
import altair as alt
import warnings
import matplotlib.pyplot as plt
warnings.simplefilter(action='ignore', category=FutureWarning)

ROOT = "https://raw.githubusercontent.com/kirenz/datasets/master/"

# ctrl + shift minus : to split content to new cell
```

```python
DATA = "loan50.csv"
df = pd.read_csv(ROOT + DATA)
```

# Notes

```python
df.info()
```

```python
df['homeownership'].value_counts().plot(kind='barh')
```

```python
alt.Chart(df).mark_bar().encode(
    x=alt.X('count()', stack='normalize'),
    y=alt.Y('homeownership').sort('-x'), 
    color='grade', 
    tooltip=['count(homeownership)','term']
).interactive()
```

```python
alt.Chart(df).mark_circle(size=60).encode(
    x='annual_income',
    y='loan_amount',
    tooltip=['annual_income', 'loan_amount', 'grade'],
    color='grade'
).properties(
    width=600,
    height=400,
    title="Scatter Plot of Annual Income vs Loan Amount"
).interactive()  
```

```python
alt.Chart(df).mark_circle(size=60).encode(
    x='annual_income',
    y='interest_rate',
    tooltip=['annual_income', 'interest_rate', 'grade'],
    color='grade'
).properties(
    width=600,
    height=400,
    title="Scatter Plot of Annual Income vs interest rate"
).interactive()  
```

```python
alt.Chart(df).mark_circle(size=60).encode(
    x='annual_income',
    y='interest_rate',
    tooltip=['annual_income', 'interest_rate', 'grade'],
    color='loan_purpose'
).properties(
    width=600,
    height=400,
    title="Scatter Plot of Annual Income vs loan purpose"
).interactive()  
```

```python
alt.Chart(df).mark_circle(size=60).encode(
    x='interest_rate',
    y='loan_purpose',
    tooltip=['annual_income', 'interest_rate', 'grade'],
    color='grade'
).properties(
    width=600,
    height=400,
    title="Scatter Plot of Annual Income vs loan purpose"
).interactive()  
```
