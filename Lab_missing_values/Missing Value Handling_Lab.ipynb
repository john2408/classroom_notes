# Missing Value Handling

```python
import pandas as pd
import missingno as msno
import seaborn as sns

from sklearn import datasets
```

# Load Dataset

```python
#load the diabetes dataset from csv into a dataframe
df = pd.read_csv('diabetes_prepared.csv')
df                   
```

# Data Exploration

```python
#show descriptive statistics
df.describe()
```

```python
#count missing values
df.isnull().sum()
```

```python
#visualize missing values
msno.matrix(df)
```

```python
#show rows with more than two columns with missing values
df[df.isnull().sum(axis=1) > 2]
```

# Missing Value Handling - Listwise Deletion

Deletion of entire row of data. If a row contains a lot of missing values and/or label is a missing.

```python
#Listwise deletion (if label is missing))
df=df.drop(df[df['label'].isnull()].index)
df
```

```python
#Listwise deletion (all rows with multiple missing values)
df=df.drop(df[df.isnull().sum(axis=1)>2].index)
df
```

```python
#visualize missing values
msno.matrix(df)
```

# Missing Value Handling - Feature Deletion

Identify a single feature with a lot of missing values. This feature shall be deleted

```python
#Delete feature x due to large number of missing values
df=df.drop('s1',axis=1)
df
```

```python
msno.matrix(df)
```

```python
#calculate correlation matrix and visualize with coolwarm color map
corr = df.corr()
sns.heatmap(corr, cmap="coolwarm", annot=True)
```

# Missing Value Handling - Average Imputation

Select a feature with missing values, which has low colinearity to other features

```python
df.head()
```

```python
from sklearn.impute import SimpleImputer
#store ids of rows with missing values in this feature

missing_values_ids_s7 = df[df['s7'].isnull()].index

#replace missing values in feature "random" with median
imputer = SimpleImputer(strategy='median')
df['s7'] = imputer.fit_transform(df[['s7']])

#show rows with missing_values_ids
df.loc[missing_values_ids_s7,:]
```

```python
msno.matrix(df)
```

# Missing Value Handling - Regression Imputation

Select the remaining feauture with missing values to be imputed with regression imputation using one or more other features

```python
#calculate correlation matrix and visualize with coolwarm color map
corr = df.corr()
sns.heatmap(corr, cmap="coolwarm", annot=True)
```

```python
from sklearn import linear_model
import numpy as np
np.random.seed(123)
reg = linear_model.LinearRegression()
#store ids of rows without missing values in feature "s3"
ids = df[df['s3'].notnull()].index.values
#train regression model
reg.fit(df.loc[ids, ['s4']], df.loc[ids, 's3'])
#show rows with missing values in feature "sepal length (cm)"
df[df['s3'].isnull()]
```

```python
#predict missing values using regression model trained on s3 and s4
df.loc[~df.index.isin(ids), 's3'] = reg.predict(df.loc[~df.index.isin(ids), ['s4']])
```
